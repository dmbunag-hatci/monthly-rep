```python
import json
import html
from pathlib import Path

IN_FILE = r"C:\Users\r103317\Projects\Monthly Report\.venv\map.json"
OUT_FILE = r"C:\Users\r103317\Projects\Monthly Report\.venv\map.json"

p = Path(IN_FILE)
if not p.exists():
    raise FileNotFoundError(f"Can't find {p.resolve()}")

with p.open("r", encoding="utf-8") as f:
    data = json.load(f)

fragments = data.get("fragments", {})
if not isinstance(fragments, dict):
    raise ValueError('Expected {"fragments": {...}}')

# Decode HTML entities in ALL values (e.g., &amp; -> &)
fragments_clean = {k: html.unescape(v) for k, v in fragments.items()}

cleaned = {"fragments": fragments_clean}

with open(OUT_FILE, "w", encoding="utf-8") as f:
    json.dump(cleaned, f, ensure_ascii=False, indent=2)

print(f"✅ Wrote cleaned map to: {OUT_FILE} ({len(fragments_clean)} entries)")
```
